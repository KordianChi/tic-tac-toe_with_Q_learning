```python
from random import randint, choice
import pandas as pd
from scipy.special import comb
from collections import Counter
from matplotlib import pyplot as plt
from jupyterthemes import jtplot
import numpy as np
from functions import control_win, activate_end, all_proper_move
from functions import teacher_player, random_player, q_learner_player
from Q_table_generator import q_table_generator
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False) # for dark mode
```

```python
# Our Q table is a dictionary where key is string with board state. Firstly we assume value 0
# for all state, except winning state.
# We generate Q table with all legal state by, significant number games between random players. 
# Number of legal state in tic tac toe is 5478, there is no obviously way to calculate that.
# Simply approach is find uppper bound of legal state.

Q_table = q_table_generator()
```
