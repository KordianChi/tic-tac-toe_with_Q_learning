```python
import numpy as np
import pandas as pd
from collections import Counter
from random import randint, choice
from matplotlib import pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
def control_win(board):
    check_board = np.array(board)
    check_board_t = np.transpose(check_board)
    for element in [check_board, check_board_t]:
        for k in range(3):
            if element[k][0] == element[k][1] and element[k][1] == element[k][2] and element[k][0]:
                return True
    if check_board[0][0] == check_board[1][1] and check_board[1][1] == check_board[2][2] and check_board[0][0]:
        return True
    elif check_board[0][2] == check_board[1][1] and check_board[1][1] == check_board[2][0] and check_board[0][2]:
        return True
    else:
        return False


def activate_end(board):
    control_list = board[0] + board[1] + board[2]
    for element in control_list:
        if element:
            active = True
        else:
            active = False
            break
    return active

def random_player(BOARD, sign):
    while True:
        row = randint(0, 2)
        column = randint(0, 2)
        if BOARD[row][column] == None:
            BOARD[row][column] = sign
            break

def teacher_player(BOARD, sign):
    if BOARD[1][1] == None:
        BOARD[1][1] = sign
        return None
    else:
        for k in [0, 2]:
            for l in [0, 2]:
                if BOARD[k][l] == None:
                    BOARD[k][l] = sign
                    return None
        for k in [0, 1, 2]:
            for l in [0, 1, 2]:
                if BOARD[k][l] == None:
                    BOARD[k][l] = sign
                    return None
```

```python
t_v_t = []
for k in range(1000):
    # game = []
    sign = 'O'
    turn = choice(['teacher_1', 'teacher_2'])
    BOARD = [[None, None, None], [None, None, None], [None, None, None]]
    while not activate_end(BOARD):
        if turn == 'teacher_1':
            teacher_player(BOARD, sign)
        else:
            teacher_player(BOARD, sign)
        for element in BOARD:
            print(element)
        print('\n')
        if control_win(BOARD):
            print(f'{turn} with {sign} win')
            # print(activate_end(BOARD))
            print(20 * '-')
            if sign == 'O':
                # game.append(BOARD[0] + BOARD[1] + BOARD[2] + [turn])
                t_v_t.append(turn)
            else:
                # game.append(BOARD[0] + BOARD[1] + BOARD[2] + [turn])
                t_v_t.append(turn)
            break
        if sign == 'O':
            sign = 'X'
        else:
            sign = 'O'
        if turn == 'teacher_1':
            turn = 'teacher_2'
        else:
            turn = 'teacher_1'
    if activate_end(BOARD) and not control_win(BOARD):
        print('Draw')
        # print(activate_end(BOARD))
        # game.append(BOARD[0] + BOARD[1] + BOARD[2] + ['D'])
        t_v_t.append('draw')
        print(20 * '-')
```

```python
r_v_t = []
for k in range(1000):
    # game = []
    sign = 'O'
    turn = choice(['random', 'teacher'])
    BOARD = [[None, None, None], [None, None, None], [None, None, None]]
    while not activate_end(BOARD):
        if turn == 'teacher':
            teacher_player(BOARD, sign)
        else:
            random_player(BOARD, sign)
        for element in BOARD:
            print(element)
        print('\n')
        if control_win(BOARD):
            print(f'{turn} with {sign} win')
            # print(activate_end(BOARD))
            print(20 * '-')
            if sign == 'O':
                # game.append(BOARD[0] + BOARD[1] + BOARD[2] + [turn])
                r_v_t.append(turn)
            else:
                # game.append(BOARD[0] + BOARD[1] + BOARD[2] + [turn])
                r_v_t.append(turn)
            break
        if sign == 'O':
            sign = 'X'
        else:
            sign = 'O'
        if turn == 'teacher':
            turn = 'random'
        else:
            turn = 'teacher'
    if activate_end(BOARD) and not control_win(BOARD):
        print('Draw')
        # print(activate_end(BOARD))
        # game.append(BOARD[0] + BOARD[1] + BOARD[2] + ['D'])
        r_v_t.append('draw')
        print(20 * '-')
```
